```python
import sqlite3

# Connect to an in-memory SQLite database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# SQL statements to create tables
sql_commands = [
    """
    CREATE TABLE Operarios (
        id_operario VARCHAR(50) PRIMARY KEY,
        nombre_completo VARCHAR(100),
        puesto_trabajo VARCHAR(50),
        fecha_incorporacion DATE,
        empresa_subcontrata VARCHAR(100) DEFAULT 'Matricerías del Centro S.L.'
    );
    """,
    """
    CREATE TABLE Biometria_Ergonomia (
        id_registro INTEGER PRIMARY KEY AUTOINCREMENT,
        id_operario VARCHAR(50) REFERENCES Operarios(id_operario),
        fecha_hora TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        indice_fatiga_actual INTEGER CHECK (indice_fatiga_actual BETWEEN 0 AND 100),
        riesgo_postural VARCHAR(20),
        alerta_predictiva_48h BOOLEAN,
        rotacion_recomendada VARCHAR(50)
    );
    """,
    """
    CREATE TABLE Aptitud_PRL (
        id_operario VARCHAR(50) REFERENCES Operarios(id_operario),
        estado_aptitud VARCHAR(50),
        fecha_ultimo_reconocimiento DATE,
        certificacion_JDS_G223 BOOLEAN,
        acceso_TKE_JDI_aprobado BOOLEAN,
        kpi_incidencias_historicas INTEGER
    );
    """,
    """
    CREATE TABLE Trazabilidad_Soberana (
        id_transaccion VARCHAR(100) PRIMARY KEY,
        id_operario VARCHAR(50) REFERENCES Operarios(id_operario),
        hash_blockchain VARCHAR(256),
        politica_acceso_ODRL VARCHAR(200),
        fecha_caducidad_dato TIMESTAMP,
        auditoria_datamatrix VARCHAR(100)
    );
    """
]

# Execute each SQL command
for command in sql_commands:
    try:
        cursor.execute(command)
    except sqlite3.Error as e:
        print(f"Error executing command: {command}\nError: {e}")

# Commit changes
conn.commit()

print("Tables created successfully in an in-memory SQLite database.")
# IMPORTANT: Do not close the connection yet, as subsequent cells will use it.
```
